```python
#留一法示例：
features = data['data']
target = data['target']
feature_names = data['feature_names']
target_names = data.target_names
labels = target_names[target]
nonSetFeatures = features[labels!='setosa']
noSetoLab = labels[labels!='setosa']
non_setosa = features[labels!='setosa']
is_virginica = noSetoLab=='virginica'
```

```python
def train_model(train_set,is_virginica):
    best_acc = -1
    for t in range(train_set.shape[1]):
        thresh = train_set[:,t]
        for j in thresh:
            prep = thresh>j
            acc = (prep == is_virginica).mean()
            if acc > best_acc:
                best_acc = acc
                best_col = t
                best_number = j
    return t,j
```

一个分类模型的构成：
1. 模型结构 ：对于本方案采用一个阈值在一个特征上的划分
2. 探索过程 ： 这里我们尽可能多的尝试所有特征和阈值的组合
3. 损失函数 ： 通过损失函数来确定哪些可能性不会太差（在机器学习领域，永远没有完美的方案）。一般希望损失函数最小化来。特别的，损失函数的构造取决于你在处理的具体问题。通常情况，是几种精力在使分类错误最少。然而，如果一些错误的代价比其他更大，也会接收一个较低的总体精度。

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors=4)
```

```python
n = len(features)
correct = 0.0
for i in range(n):
    training = np.ones(n,bool)
    training[i] = 0
    testing = ~training
    classifier.fit(features[training],labels[training])
    pred = classifier.predict(features[testing])
    correct += (pred == labels[i])
print('Result of leave-one-out: {}'.format(correct/n))
```

采用交叉验证进行准确率的预测

```python
from sklearn.cross_validation import KFold
```

```python
means = []
kf = KFold(len(features),n_folds=3,shuffle=True)
for training,testing in kf:
    classifier.fit(features[training],labels[training])
    prediction = classifier.predict(features[testing])
    
    curmean = np.mean(prediction == labels[testing])
    means.append(curmean)
print('Result of cross-validation using KFold: {}'.format(means))
```

```python
def load_dataset(dataset_name):
    data = []
    labels = []
    with open('{0}.tsv'.format(dataset_name)) as ifile:
        for line in ifile:
            tokens = line.strip().split('\t')
            data.append([float(tk) for tk in tokens[:-1]])
            labels.append(tokens[-1])
    data = np.array(data)
    labels = np.array(labels)
    return data, labels
```
